```python
import requests
from bs4 import BeautifulSoup

url = "https://getbootstrap.com/docs/5.0/getting-started/download/"
r = requests.get(url)


soup = BeautifulSoup(r.text,"lxml")
tags = soup.h1
all_tags = soup.body
att = tags.attrs
print("All the tags on this page: ", all_tags)
print(att)
print(tags.string,f"and it is Containing the atts: {att}")
```

```python
# Now using the find function we can narrow down the scrapping

tags = soup.find("attrs")
tags
```

```python
title = soup.find("h2")
title.string
```

```python
div_tag = soup.find('div', class_='bd-intro')
print(div_tag)
```
