```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
re = requests.get(url)
re.status_code
```

```python
re.text
```

```python
soup  = BeautifulSoup(re.text,"lxml")
soup
```

```python
att = soup.find("attrs")
att
```

```python
title =  soup.find("a",{"class":"title"})
title.string
```

```python
price = soup.find("h4",{"class":"price"})
price.string
```

```python
des = soup.find("p",{"class":"description"})
des.string
```
